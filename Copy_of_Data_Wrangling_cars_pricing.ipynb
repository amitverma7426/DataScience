```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#Here's how it's calculated:

# Minimum value of "horsepower" column: Let's say it's 48.
# Maximum value of "horsepower" column: Let's say it's 288.
# Number of bins: 4 (as specified in the third argument of np.linspace()).
# The function then divides the interval between the minimum and maximum values into 3 equal segments (since you want 4 bins), including both the minimum and maximum values.

# So, the interval width between each bin edge is (288 - 48) / (4 - 1) = 80.

# Starting from the minimum value (48), the bins are generated at intervals of 80:

# First bin edge: 4
# Second bin edge: 48 + 80 = 128
# Third bin edge: 128 + 80 = 208
# Fourth bin edge: 208 + 80 = 288
```

```python
#Set group  names:

group_names = ['Low', 'Medium', 'High']
group_names
```

```python
#Apply the function "cut" to determine what each value of df['horsepower'] belongs to
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#See the number of vehicles in each bin:
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
