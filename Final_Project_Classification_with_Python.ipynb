```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv'
```

```python
df = pd.read_csv(path)
df.head()
```


1.Data prepocesssion

```python
 df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
df_sydney_processed
```

```python
df.columns
```

```python
# Change df.columns() to df.columns
df.columns
```

```python
  df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```



2. Training Data and Test Data



```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
df_sydney_processed
```

```python
X = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```



3. Linear Regression



```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
LinearReg = LinearRegression()
LinearReg.fit(X_train, Y_train)
```

```python
predictions = LinearReg.predict(X_test)
predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate Mean Absolute Error
LinearRegression_MAE = mean_absolute_error(Y_test, predictions)

# Calculate Mean Squared Error
LinearRegression_MSE = mean_squared_error(Y_test, predictions)

# Calculate R-squared
LinearRegression_R2 = r2_score(Y_test, predictions)

# Print the calculated metrics
print("Linear Regression Metrics:")
print("Mean Absolute Error:", LinearRegression_MAE)
print("Mean Squared Error:", LinearRegression_MSE)
print("R-squared:", LinearRegression_R2)
```

```python
# Create a DataFrame to store the metrics
Report = pd.DataFrame({
    'Metric': ['Mean Absolute Error (MAE)', 'Mean Squared Error (MSE)', 'R-squared (R2)'],
    'Value': [LinearRegression_MAE, LinearRegression_MSE, LinearRegression_R2]
})

# Display the DataFrame
print("Linear Regression Metrics:")
print(Report)
```



4.**Logistic** **Regression**



```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,Y_train)
LR
```

```python
predictions = LR.predict(X_test)
predictions
```

```python
predict_proba = LR.predict_proba(X_test)
predict_proba
```

```python
#LR_Accuracy_Score =
#LR_JaccardIndex =
#LR_F1_Score =
#LR_Log_Loss =
```

```python
from sklearn.metrics import accuracy_score, jaccard_score, f1_score, log_loss
LR_Accuracy_Score = accuracy_score(Y_test, predictions,)
LR_JaccardIndex = jaccard_score(Y_test, predictions)
LR_F1_Score = f1_score(Y_test, predictions)
LR_Log_Loss = log_loss(Y_test, predict_proba)
print("Logistic Regression Accuracy Score:", LR_Accuracy_Score)
print("Logistic Regression Jaccard Index:", LR_JaccardIndex)
print("Logistic Regression F1 Score:", LR_F1_Score)
print("Logistic Regression Log Loss:", LR_Log_Loss)
```

```python
# Create a DataFrame to store the metrics
Report4 = pd.DataFrame({
    'Metric': ['Logistic Regression Accuracy Score', 'Logistic Regression Jaccard Index', 'Logistic Regression F1 Score', 'Logistic Regression Log Loss'],
    'Value': [  LR_Accuracy_Score, LR_JaccardIndex,  LR_F1_Score,LR_Log_Loss ]
})
Report4
```
