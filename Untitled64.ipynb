```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# Define the Logic Gate dataset
# X: input data, Y: expected output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [0]])

# Define the model
model = keras.Sequential([
    layers.Dense(2, input_shape=(2,), activation='sigmoid'),
    layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X, Y, epochs=5000)

# Predict the output for the input data
predictions = model.predict(X)

# Print the results
print("Predictions:", predictions)

# Visualize the decision boundary
x1 = np.arange(0, 2, 0.01)
x2 = np.arange(0, 2, 0.01)
X1, X2 = np.meshgrid(x1, x2)
predictions = model.predict(np.array([X1.ravel(), X2.ravel()]).T)
plt.contourf(X1, X2, predictions.reshape(X1.shape), cmap='RdBu')
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='RdBu')
plt.show()
```
